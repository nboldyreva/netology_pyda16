# Задание 1.

Напишите функцию, которая принимает на вход строку и проверяет является ли она валидным транспортным номером. Если да, то функция должна возвращать отдельно номер и регион.

```python
def format_grz(raw_gosnomer):
    regexp = r'^([АВЕКМНОРСТУХ]\d{3}[АВЕКМНОРСТУХ]{2})(\d{1,3})$'
    result = re.findall(regexp,raw_gosnomer,re.IGNORECASE)
    if len(result) > 0:
        return result[0]
    else:
        return False
    
result = format_grz(input('Введите государственный регистрационный номер (буквы кириллицей): '))
if result:
    print(f'Номер: {result[0].upper()}\nРегион: {result[1]}')
else:
    print('Некорректный формат ГРЗ')
```

# Задание 3.
Напишите функцию, которая будет удалять все последовательные повторы слов из заданной строки при помощи регулярных выражений.

```python
def remove_duplicates (input_string):
    return re.sub(r'(\w+)\s+\1([\.\,\?\!\-\:\;\s])?', r'\1\2', input_string)

remove_duplicates(input("Введите строку с повторяющимися словами: "))
```

# Задание 4.
Напишите функцию, которая будет проверять номер сотового телефона на валидность, если он валиден, то переводить его в формат:
+7(xxx)-xxx-xx-xx

Постарайтесь предусмотреть как можно больше адекватных форматов изначального ввода номера. Примеры валидных форматов:

* +7 955 555-55-55;
* 9555555555;
* 8(955)555-55-55;
* +7 955 555 55 55;
* 7(955) 555-55-55;
* +7 955+555+55+55.

```python
def format_phone(raw_number):
    regexp = r'[\-\+\*]?(7|8)?[\s\-\+\*]?\(?(\d{3})\)?[\s\-\+\*]?(\d{3})[\s\-\+\*]?(\d{2})[\s\-\+\*]?(\d{2})'
    return re.sub(regexp, r'+7(\2)-\3-\4-\5', raw_number)

format_phone(input("Введите номер (10 или 11 цифр): "))
```
